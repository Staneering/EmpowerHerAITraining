# **Assignment: Secret Code Translator**  
**Difficulty**: Beginner  
**Objective**: Practice string manipulation, dictionaries, and user input in Python by building a vowel-to-symbol cipher.

---

### **Problem Statement**  
Create a program that converts a user's message into a "secret code" by replacing all vowels with predefined symbols. The program should:  
1. Ask the user to input a message.  
2. Replace each vowel (a, e, i, o, u) with a corresponding symbol.  
3. Print the encoded message.  

---

### **Requirements**  
#### **1. Symbol Mapping**  
Use the following replacements (case-insensitive):  
- `a` → `@`  
- `e` → `#`  
- `i` → `$`  
- `o` → `%`  
- `u` → `&`  

#### **2. User Interaction**  
- Prompt the user with: `"Enter your message: "`.  
- Display the encoded message with: `"Secret code: [RESULT]"`.  

#### **3. Example Outputs**  
- **Input**: `Hello World`  
  **Output**: `H#ll% W%rld`  
- **Input**: `Python is fun!`  
  **Output**: `Pyth%n $s f&n!`  

---

### **Key Concepts to Use**  
1. **String Methods**:  
   - Iterate through each character in the input string.  
   - Check if a character is a vowel (case-insensitive).  
2. **Dictionaries**:  
   - Store vowel-to-symbol mappings efficiently.  
3. **Conditionals**:  
   - Decide whether to replace a character or keep it.  

---

### **Hints (Optional Section)**  
- Use a dictionary to map vowels to symbols.  
- Convert the input to lowercase (or uppercase) to simplify comparisons.  
- Build the result character-by-character in a loop.  

---

### **Bonus Challenges**  
1. **Case Sensitivity**:  
   Preserve the original case in the output (e.g., `E` → `#`, `e` → `#`).  
2. **Reverse Translator**:  
   Add a feature to decode the secret code back to the original message.  
3. **Custom Symbols**:  
   Let the user define their own vowel-to-symbol mappings.  

---

### **Submission Guidelines**  
1. Name your file: `secret_code.py`.  
2. Include comments to explain your logic.  
3. Test your code with the provided examples and edge cases (e.g., no vowels, mixed case).
