## üß† What is a String?
# A string is a sequence of characters enclosed in single (' '), double (" "), or triple quotes (''' ''' or """ """).

# Example:
greeting = "Hello, World!"

# ---

## üõ†Ô∏è Common String Methods

sample = "  Python Programming is fun!  "

### 1. strip()
# Removes leading and trailing whitespaces
print("Strip:", sample.strip())

### 2. lower()
# Converts the string to lowercase
print("Lower:", sample.lower())

### 3. upper()
# Converts the string to uppercase
print("Upper:", sample.upper())

### 4. title()
# Capitalizes the first letter of each word
print("Title:", sample.title())

### 5. replace()
# Replaces occurrences of a substring with another
print("Replace:", sample.replace("fun", "awesome"))

### 6. split()
# Splits the string into a list based on a delimiter (default is space)
print("Split:", sample.split())

### 7. join()
# Joins elements of an iterable with the string as separator
words = ["Python", "is", "cool"]
print("Join:", "-".join(words))

### 8. find()
# Returns the index of the first occurrence of a substring (or -1 if not found)
print("Find 'gram':", sample.find("gram"))

### 9. startswith()
# Checks if the string starts with a specific substring
print("Starts with '  Python':", sample.startswith("  Python"))

### 10. endswith()
# Checks if the string ends with a specific substring
print("Ends with 'fun!  ':", sample.endswith("fun!  "))

### 11. count()
# Returns the number of times a substring appears
print("Count of 'm':", sample.count("m"))

### 12. isalpha()
# Checks if all characters are alphabets (no spaces or numbers)
alpha = "Hello"
print("Is alpha:", alpha.isalpha())

# ---

## üîÅ String Slicing
# Like lists, strings support slicing: string[start:stop:step]
text = "Programming"
print("First 5 chars:", text[:5])
print("Every second char:", text[::2])
print("Reversed:", text[::-1])
