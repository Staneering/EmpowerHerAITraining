### **1. The Shopping List Simulator**  
**Task**: Create a program where users can:  
- Add items to a virtual shopping list.  
- Remove items they've "bought".  
- Print the list sorted alphabetically.  
**Concept**: List methods (`append()`, `remove()`, `sort()`).

---

### **2. Treasure Hunt (Index Challenge)**  
**Task**:  
- Hide a "treasure" (e.g., the number `42`) in a random position in a list.  
- Students write code to find its index.  
**Concept**: List indexing, `list.index()`.

### **4. The Never-Ending To-Do List**  
**Task**:  
- Use a `while` loop to let users keep adding tasks until they type "done".  
- Bonus: Let them mark tasks as complete (move to a `done_tasks` list).  
**Concept**: Infinite loops, nested lists.

### **5. List Roulette (Random Choice)**  
**Task**:  
- Create a list of funny dares (e.g., `["Sing a song", "Do 10 pushups"]`).  
- Use `random.choice()` to pick one.  
**Concept**: Importing modules, randomness.

### **6. The Alphabet Game**  
**Task**:  
- Generate a list of letters from `A-Z` using `list(range(65, 91))` and `chr()`.  
- Challenge: Can students decode `[72, 69, 76, 76, 79]` into a word?  
**Concept**: ASCII conversion, list comprehensions.

### **7. Duplicate Detective**  
**Task**:  
- Provide a list with duplicates (e.g., `[5, 2, 5, 8]`).  
- Students write code to find/remove duplicates.  
**Concept**: Conditionals, `in` operator.

### **8. The Sentence Mixer**  
**Task**:  
- Split a sentence into words with `split()`.  
- Shuffle the list and print it as a "mad libs" sentence.  
**Concept**: String splitting, `random.shuffle()`.

### **10. The Password Generator**  
**Task**:  
- Combine lists of adjectives, nouns, and numbers (e.g., `["happy", "cat", "123"]`).  
- Randomly generate a password like `"happycat123"`.  
**Concept**: List concatenation, `random.sample()`.
