## **Fundamental Concepts**  

### **1. Variables & Data Types**  
```python
name = "Alice"      # String (text)  
age = 25           # Integer (whole number)  
height = 5.9       # Float (decimal)  
is_student = True  # Boolean (True/False)  
```

### **2. Basic Operations**  
```python
# Math  
sum = 3 + 2    # 5  
product = 3 * 2  # 6  

# String concatenation  
greeting = "Hello, " + name  # "Hello, Alice"  
```

### **3. Control Flow**  
- **Conditionals (`if-else`)**  
  ```python
  if age >= 18:  
      print("Adult")  
  else:  
      print("Minor")  
  ```
- **Loops (`for`, `while`)**  
  ```python
  for i in range(3):  # Prints 0, 1, 2  
      print(i)  
  ```

### **4. Functions**  
```python
def greet(name):  
    return f"Hello, {name}!"  

print(greet("Bob"))  # "Hello, Bob!"  
```

### **5. Lists & Dictionaries**  
```python
fruits = ["apple", "banana", "cherry"]  # List (ordered)  
person = {"name": "Alice", "age": 25}   # Dictionary (key-value pairs)  
```
