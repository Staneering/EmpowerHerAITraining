

### **"Reversible Cipher Generator"**  

ðŸ”¹ **Task**:  
Create a program that **encodes** and **decodes** a message using a simple cipher (like shifting letters by a fixed number).  

ðŸ”¹ **Cipher Rules**:  
- **Encode**: Shift each letter in the message *forward* by `n` in the alphabet (e.g., `n=3`: `A â†’ D`, `B â†’ E`).  
- **Decode**: Shift letters *backward* by `n` to reverse it.  
- Non-alphabet characters (spaces, numbers, symbols) remain unchanged.  

ðŸ”¹ **Requirements**:  
1. Ask the user if they want to **encode** or **decode**.  
2. Take their message and shift value (`n`).  
3. Transform the message and display the result.  

ðŸ”¹ **Example**:  
**Input (Encode, n=3)**:  
`Hello, World!`  

**Output**:  
`Khoor, Zruog!`  

**Input (Decode, n=3)**:  
`Khoor, Zruog!`  

**Output**:  
`Hello, World!`

### **Step-by-Step Algorithm**  
1. **Ask the user**:  
   ```python
   mode = input("Encode (e) or Decode (d)? ").lower()  
   n = int(input("Shift value (n): "))  
   message = input("Enter message: ")  
   ```
2. **Shift letters**:  
   - For each character:  
     - If itâ€™s a letter, calculate its new position (use `ord()` and `chr()`).  
     - Handle wrap-around (e.g., `z + 3 â†’ c`).  
     - Non-letters stay the same.  
3. **Display result**.  

---

### **Sample Code**  
```python
def cipher(message, shift, mode):  
    result = ""  
    for char in message:  
        if char.isalpha():  
            base = ord('a') if char.islower() else ord('A')  
            shifted = ord(char) - base  
            if mode == 'e':  # Encode  
                new_pos = (shifted + shift) % 26  
            else:  # Decode  
                new_pos = (shifted - shift) % 26  
            result += chr(base + new_pos)  
        else:  
            result += char  
    return result  

# User input  
mode = input("Encode (e) or Decode (d)? ").lower()  
shift = int(input("Shift value (n): "))  
message = input("Enter message: ")  

# Output  
print("Result:", cipher(message, shift, mode))  
```

### **Caesar Cipher Algorithm**
**Inputs**:  
- `message`: String to encode/decode  
- `shift`: Integer (number of positions to shift)  
- `mode`: 'e' for encode, 'd' for decode  

**Output**: Encoded/decoded string  

---

#### **Steps**:  
1. **Initialize** an empty string `result` to store the output.  
2. **Loop** through each character (`char`) in the input `message`:  
   - **If** the character is alphabetic (`char.isalpha()`):  
     a. **Determine the base value**:  
        - `ord('a')` for lowercase letters.  
        - `ord('A')` for uppercase letters.  
     b. **Calculate the shifted position**:  
        - Convert `char` to its ASCII value (`ord(char)`).  
        - Subtract the base to get a 0-based index (`shifted = ord(char) - base`).  
     c. **Apply the shift**:  
        - **Encode (`mode == 'e'`)**:  
          ```python
          new_pos = (shifted + shift) % 26  # Wrap around after 'z'/'Z'
          ```  
        - **Decode (`mode == 'd'`)**:  
          ```python
          new_pos = (shifted - shift) % 26  # Wrap around before 'a'/'A'
          ```  
     d. **Convert back to a character**:  
        - Add `new_pos` to the base and convert to a character (`chr(base + new_pos)`).  
     e. **Append** to `result`.  
   - **Else** (non-alphabetic character):  
     - Append the character to `result` unchanged.  
3. **Return** the final `result` string.

### **Example Walkthrough**  
**Input**:  
```python
cipher("Hello", 3, 'e')
```  

**Execution**:  
1. `H` â†’ Uppercase â†’ Base = `ord('A')` (65)  
   - `shifted = ord('H') - 65 = 7`  
   - `new_pos = (7 + 3) % 26 = 10`  
   - `result += chr(65 + 10) = 'K'`  
2. `e` â†’ Lowercase â†’ Base = `ord('a')` (97)  
   - `shifted = ord('e') - 97 = 4`  
   - `new_pos = (4 + 3) % 26 = 7`  
   - `result += chr(97 + 7) = 'h'`  
3. Repeat for `'l'`, `'l'`, `'o'` â†’ `'o'` becomes `'r'`.  
4. **Final Output**: `"Khoor"`  

---

### **Key Features**:  
- **Preserves case**: Lowercase â†’ lowercase, uppercase â†’ uppercase.  
- **Handles wrap-around**: Shifting `'z'` by 3 becomes `'c'`.  
- **Skips non-alphabetic**: Numbers/symbols remain unchanged.
